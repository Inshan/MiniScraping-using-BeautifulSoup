```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
r= requests.get('https://en.wikipedia.org/wiki/Cillian_Murphy')
```

```python
soup = bs(r.content)
print(soup.prettify())
```

```python
links = soup.select('p')
links
```

```python
for link in links:
    print(link.get_text())
```

```python
links = soup.select('h1')
for link in links:
    print(link.get_text())
```

```python
links = soup.select('h2')
for link in links:
    print(link.get_text())
```

```python
links = soup.select('h3')
for link in links:
    print(link.get_text())
```

```python
paragraphs = soup.select('p')

for paragraph in paragraphs:
    links_in_paragraph = paragraph.select('a')
    for link in links_in_paragraph:
        link_text = link.get_text()
        if not link_text.startswith('['):
            print(link_text)
```
